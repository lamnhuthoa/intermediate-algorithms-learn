# Transform the Expression

Transform an algebraic expression containing brackets into **RPN form** (Reverse Polish Notation).

### Rules:
* **Two-argument operators:** `+`, `-`, `*`, `/`, `^` (priority from lowest to highest: `+` < `-` < `*` < `/` < `^`).
* **Brackets:** `(` and `)`.
* **Operands:** Only lowercase letters `a, b, ..., z`.
* **Assumption:** There is only one unique RPN form for each expression (no ambiguous cases like `a*b*c`).

### Input Format:
* The first line contains $t$ (the number of expressions, $t \le 100$).
* Each of the next $t$ lines contains an expression with length $\le 400$.

### Output Format:
* For each expression, print its RPN form on a new line.

**Giải thích ví dụ: `(a+(b*c))`**

1. Gặp `(`: Stack = `[` `(` `]`
2. Gặp `a`: Kết quả = `a`
3. Gặp `+`: Stack = `[` `(`, `+` `]`
4. Gặp `(`: Stack = `[` `(`, `+`, `(` `]`
5. Gặp `b`: Kết quả = `ab`
6. Gặp `*`: Stack = `[` `(`, `+`, `(`, `*` `]`
7. Gặp `c`: Kết quả = `abc`
8. Gặp `)` đầu tiên: Lấy `*` ra. Kết quả = `abc*`. Stack còn `[` `(`, `+` `]`
9. Gặp `)` cuối cùng: Lấy `+` ra. Kết quả = `abc*+`. Stack trống. Kết quả cuối cùng: `abc*+`
