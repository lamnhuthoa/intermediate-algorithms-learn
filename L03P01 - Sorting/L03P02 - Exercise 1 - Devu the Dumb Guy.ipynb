# Devu, the Dumb Guy

Devu is a dumb guy, his learning curve is very slow. You are supposed to teach him $n$ subjects, the $i$-th subject has $c_i$ chapters. When you teach him, you must teach all chapters of a subject continuously.

### Rules:
* His initial learning time per chapter is $x$ hours.
* After teaching him a full subject, the time required for any chapter of the **next** subject decreases by exactly **1 hour**.
* The learning time per chapter **cannot be less than 1 hour**.
* You can teach the $n$ subjects in **any order**.

### Task:
Find the **minimum** amount of time (in hours) Devu will take to understand all subjects.

### Input Format:
- The first line contains two integers $n$ and $x$ ($1 \le n, x \le 10^5$).
- The second line contains $n$ integers $c_1, c_2, \dots, c_n$ ($1 \le c_i \le 10^5$) â€” the number of chapters in each subject.

### Output Format:
- Output a single integer representing the minimum total time. (Note: The answer may exceed 32-bit integer limits).
