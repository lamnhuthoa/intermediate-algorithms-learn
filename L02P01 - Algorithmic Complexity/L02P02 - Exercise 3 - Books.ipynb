# Books

When Valera has some free time, he goes to the library to read. Today, he has exactly **$t$** free minutes. 

Valera has chosen **$n$** books in the library and numbered them from $1$ to $n$. For each book $i$, he estimated the time needed to read it as **$a_i$** minutes.

### Valera's Reading Rules:
1. Valera chooses an arbitrary starting book $i$.
2. He reads the books **one by one** in order: first book $i$, then book $i+1$, then book $i+2$, and so on.
3. The process continues until:
    * He finishes reading the last book ($n^{th}$ book).
    * **Or** he runs out of free time and cannot finish the next book.
4. Valera only reads books to the end. He will **not start** reading a book if he does not have enough time to finish it completely.

### Task:
Find the **maximum number of books** Valera can read.

---

### Input Format:
* **The first line** contains two integers: $n$ ($1 \le n \le 10^5$) and $t$ ($1 \le t \le 10^9$) — the number of books and the total free minutes Valera has.
* **The second line** contains a sequence of $n$ integers $a_1, a_2, \dots, a_n$ ($1 \le a_i \le 10^4$), where $a_i$ is the time needed to read the $i^{th}$ book.

### Output Format:
* Print a single integer — the maximum number of books Valera can read.

---

### Sample Tests:

| Input | Output | Explanation |
| :--- | :--- | :--- |
| `4 5`<br>`3 1 2 1` | `3` | Start from book 2: $1+2+1 = 4 \le 5$. Total 3 books. |
| `3 3`<br>`2 2 3` | `1` | Any single book is $\le 3$, but any two together exceed 3. |

---

### Complexity Analysis:
* **Time Complexity:** $O(n)$ should be targeted using a **Sliding Window** or **Two Pointers** approach.
* **Space Complexity:** $O(n)$ to store the array of reading times.
