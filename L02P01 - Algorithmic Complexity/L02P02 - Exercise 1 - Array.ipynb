# Find Minimum Segment with k Distinct Values

Given an array $a$ of $n$ elements: $a_1, a_2, \dots, a_n$.

### Task:
Find a **minimal** segment $[l, r]$ ($1 \le l \le r \le n$) such that among the elements $a_l, a_{l+1}, \dots, a_r$, there are exactly $k$ distinct values.

A segment $[l, r]$ is called **minimal** if there is no segment $[x, y]$ that also satisfies the property (having $k$ distinct values) and is contained within $[l, r]$ with a smaller length, where $1 \le l \le x \le y \le r \le n$.

Note: The segment $[l, r]$ does not need to be the absolute shortest among all possible segments; it just needs to be "minimal" (cannot be shrunk further while keeping $k$ distinct values).

### Input Format:
- Line 1: Two integers $n$ and $k$ ($1 \le n, k \le 10^5$).
- Line 2: $n$ integers $a_1, a_2, \dots, a_n$ ($1 \le a_i \le 10^5$).

### Output Format:
- Print two integers $l$ and $r$. If no such segment exists, print `-1 -1`. 
- If multiple answers exist, any one is acceptable.

| Input Case | Diễn giải thuật toán | Trạng thái cửa sổ tối ưu | Output |
| :--- | :--- | :--- | :--- |
| `4 2` <br> `1 2 2 3` | `right` dừng tại index 1 (giá trị 2). `unique=2`. Không thể thu hẹp `left`. | `[1, 2]` | `1 2` |
| `8 3` <br> `1 1 2 2 3 3 4 5` | `right` dừng tại index 4 (giá trị 3). `unique=3`. `left` thu hẹp từ 0 lên 1 vì `a[0]=1` dư thừa. | `[1, 2, 2, 3]` (index 1 đến 4) | `2 5` |
| `7 4` <br> `4 7 7 4 7 4 7` | Chỉ có 2 giá trị {4, 7}. `unique` không bao giờ đạt tới 4. | Không có | `-1 -1` |

**Giải thích Case 2:**
1. Cửa sổ mở rộng đến `[1, 1, 2, 2, 3]` (từ index 0 đến 4). Lúc này `unique_count = 3`.
2. Phần tử `a[0] = 1` xuất hiện 2 lần trong cửa sổ, nên ta dịch `left` lên 1 đơn vị.
3. Cửa sổ còn lại `[1, 2, 2, 3]` (index 1 đến 4). `a[1] = 1` chỉ còn 1 lần, không thể thu hẹp thêm.
4. Kết quả 1-indexed: `2 5`.
